# NCEP GODAS

```python
# https://downloads.psl.noaa.gov//Datasets/godas/ucur.1980.nc
# https://downloads.psl.noaa.gov//Datasets/godas/vcur.1980.nc
# https://downloads.psl.noaa.gov//Datasets/godas/pottmp.1980.nc
# https://downloads.psl.noaa.gov//Datasets/godas/dbss_obml.1980.nc

f=open('godas.sh','w')
for i in range(1982,2021):
    #print(f"curl  -C - https://downloads.psl.noaa.gov//Datasets/godas/ucur.{i}.nc --output godas/u/u_{i}.nc")
    f.write(f"curl  -C - https://downloads.psl.noaa.gov//Datasets/godas/ucur.{i}.nc --output godas/u/u_{i}.nc")
    f.write('\n')
    f.write(f"curl  -C - https://downloads.psl.noaa.gov//Datasets/godas/vcur.{i}.nc --output godas/v/v_{i}.nc")
    f.write('\n')
    f.write(f"curl  -C - https://downloads.psl.noaa.gov//Datasets/godas/pottmp.{i}.nc --output godas/sst/pottmp_{i}.nc")
    f.write('\n')
    f.write(f"curl  -C - https://downloads.psl.noaa.gov//Datasets/godas/dbss_obml.{i}.nc --output godas/mld/mld_{i}.nc")
    f.write('\n')
f.close()
```

# Ncep Reanalysis

```python
# https://downloads.psl.noaa.gov//Datasets/ncep.reanalysis/Dailies/surface_gauss/nswrs.sfc.gauss.1950.nc
f=open('ncep.sh','w')
for i in range(1951,2021):
    #print(f"curl  -C - https://downloads.psl.noaa.gov//Datasets/godas/ucur.{i}.nc --output godas/u/u_{i}.nc")
    f.write(f"curl  -C - https://downloads.psl.noaa.gov//Datasets/ncep.reanalysis/Dailies/surface_gauss/nswrs.sfc.gauss.{i}.nc --output ncar/SW/SW_{i}.nc")
    f.write('\n')
    f.write(f"curl  -C - https://downloads.psl.noaa.gov//Datasets/ncep.reanalysis/Dailies/surface_gauss/nlwrs.sfc.gauss.{i}.nc --output ncar/LW/LW_{i}.nc")
    f.write('\n')
f.close()
```
