```python
t=pd.date_range('19930101','20221231',freq='M')
ts=pd.date_range('19930101','20221231',freq='D')
all=[]
for i in t:
    all.append([str(i.year),f'{i.month:02d}','1'])
    all.append([str(i.year),f'{i.month:02d}','2'])
    all.append([str(i.year),f'{i.month:02d}','3'])
    
def justncnpy(s):
    allows=['npy','nc','NPY','NC']
    if s.split('.')[-1] in allows:
        return True
    else:
        return False      
def getda(s):
    ks=s.split('.')[0].split('_')[2:]
    if len(ks)==2:
        a1=s.split('.')[0].split('_')[2:];a1.append('1')
        a2=s.split('.')[0].split('_')[2:];a2.append('2')
        a3=s.split('.')[0].split('_')[2:];a3.append('3')
        return a1,a2,a3
    elif len(ks)==3:
        return ks
def sets(s):
    a=''
    for i in s:
        a+=i
        a+='_'
    return a[:-1]
def setss(li):
    k=list(map(sets,li))
    k.sort()
    return set(k)
    
temps=[];temps.extend(temps_loc);temps.extend(temps_ssh)
temps=np.unique(np.array(temps)[list(map(justncnpy,temps))])


us=[];us.extend(u_loc);us.extend(u_ssh)
us=np.unique(np.array(us)[list(map(justncnpy,us))])

vs=[];vs.extend(v_loc);vs.extend(v_ssh)
vs=np.unique(np.array(vs)[list(map(justncnpy,vs))])


tmp=[]
for i in list(map(getda,temps)):
    if type(i)==tuple:
        tmp.append(i[0]);tmp.append(i[1]);tmp.append(i[2])
    else:
        tmp.append(i)
temps=tmp

tmp=[]
for i in list(map(getda,us)):
    if type(i)==tuple:
        tmp.append(i[0]);tmp.append(i[1]);tmp.append(i[2])
    else:
        tmp.append(i)
us=tmp

tmp=[]
for i in list(map(getda,vs)):
    if type(i)==tuple:
        tmp.append(i[0]);tmp.append(i[1]);tmp.append(i[2])
    else:
        tmp.append(i)
vs=tmp

temp_add=list(setss(all) - setss(temps))
u_add=list(setss(all) - setss(us))
v_add=list(setss(all) - setss(vs))
temp_add.sort()
u_add.sort()
v_add.sort()
```
