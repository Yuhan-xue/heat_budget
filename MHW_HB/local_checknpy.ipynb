```python
import matplotlib.pyplot as plt
from os import listdir
import netCDF4 as nc
import numpy as np
import pandas as pd
import datetime
```

```python
## 获取文件夹里所有NC文件
def getfiles(filed):
    files=listdir(filed)
    files.sort()
    files=files[0:]
    files_n=[filed+'/'+i for i in files]
    return files_n
```

```python
npyfile=getfiles(r"F:\OceanData\Bran\temp3")
for i in npyfile:
    data=np.load(i)
    ts=np.nanmean(np.nanmean(np.nanmean(data,axis=0),axis=-1),axis=-1)
    print(i,end='  ')
    print(ts[ts>=80],end='  ')
    print(ts)
```
