```python
import datetime
import pandas as pd
import netCDF4 as nc
import numpy as np
import matplotlib.pyplot as plt
from os import listdir
from scipy.interpolate import RegularGridInterpolator
from concurrent.futures import ThreadPoolExecutor, ProcessPoolExecutor, Executor
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
## 从NC文件时间转PD时间
def cftime2pdtime(cf):
    return pd.to_datetime(datetime.datetime(cf.year,cf.month,cf.day,cf.hour))

## 获取文件夹里所有NC文件
def getfiles(filed):
    files=listdir(filed)
    files.sort()
    files=files[0:]
    files_n=[filed+'/'+i for i in files]
    return files_n

def wr(txt):
    w=open('日志.log','a')
    w.write(txt)
    w.close()
```

```python
das1=nc.Dataset('/lustre/home/yuhanxue/data/BRAN/ERA5_4Q/1993_150E_180.nc')
das2=nc.Dataset('/lustre/home/yuhanxue/data/BRAN/ERA5_4Q/1993_0_70W.nc')
lon1=np.array(das1['longitude'])
lon2=np.array(das2['longitude'])
lat_e=np.array(das1['latitude'])
lon_e=np.concatenate([lon1[:-1],lon2+360])
lon_e=np.arange(np.min(lon_e),np.max(lon_e)+1,2)
lat_e=np.arange(np.min(lat_e), np.max(lat_e)+1, 2)
```

```python

def getnc(fi):
    if fi.split('.')[-1]=='nc':
        return True
    else:
        return False

file_add='/lustre/home/yuhanxue/data/BRAN/mld'
files=np.array(getfiles(file_add))[list(map(getnc,getfiles(file_add)))]
das=nc.Dataset(files[0])

lon_4=lon_e
lat_4=lat_e
lon_12=np.array(das['xt_ocean'])
lat_12=np.array(das['yt_ocean'])
Lon_4,Lat_4=np.meshgrid(lon_4,lat_4)
wr(f'lat_4 range:{lat_4[0]:.2f}~{lat_4[-1]:.2f} | freq:{abs(lat_4[1]-lat_4[0])}\nlon_4 range:{lon_4[0]:.2f}~{lon_4[-1]:.2f} | freq:{lon_4[1]-lon_4[0]}')
wr(f'lat_12 range:{lat_12[0]:.2f}~{lat_12[-1]:.2f} | freq:{abs(lat_12[1]-lat_12[0])}\nlon_12 range:{lon_12[0]:.2f}~{lon_12[-1]:.2f} | freq:{lon_12[1]-lon_12[0]}')
def grid(dat):
    global lon_12,lat_12,Lon_4,Lat_4
    #wr(dat.shape)
    #f=RegularGridInterpolator((lat_12,lon_12),godas_mlds[1,:,:],bounds_error=False,method='qubic',fill_value=np.nan)
    return RegularGridInterpolator((lat_12,lon_12),dat,bounds_error=False,fill_value=np.nan)((Lat_4,Lon_4))
def allinone(dat):
    pool = ThreadPoolExecutor(max_workers=4)
    ans=np.array(list(pool.map(grid,dat)))
    del pool
    return ans

def readnc(fi):
    das=nc.Dataset(fi)
    data=np.array(das['mld'])
    data[data<-200]=np.nan
    return allinone(data)
```
